# Annotated Plot

```python
x = np.arange(0, 20)
y = x ^ 3

# Plotting the graph
plt.plot(x, y)

# Adding annotations
plt.annotate('Start', xy=(x[0], y[0]), xytext=(x[0] + 2, y[0] + 2),
             arrowprops=dict(facecolor='red', arrowstyle='->'))
plt.annotate('End', xy=(x[-1], y[-1]), xytext=(x[-1] - 5, y[-1] - 2),
             arrowprops=dict(facecolor='blue', arrowstyle='->'))

# Optional: Add labels and title for clarity
plt.title('Annotated Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plot
plt.show()
```

## Plot with Distance Annotations

```python
x = np.arange(0, 20)
y = x ^ 3

# Plotting the graph
plt.plot(x, y)

# Mark two points to calculate the "distance"
point1 = (x[5], y[5])  # Example: 5th point
point2 = (x[15], y[15])  # Example: 15th point

# Add annotations to mark the points
plt.scatter(*point1, color='red')  # Point 1
plt.scatter(*point2, color='blue')  # Point 2

# Draw a line between the points to show the distance
plt.plot([point1[0], point2[0]], [point1[1], point2[1]], 'k--', label='Distance Line')

# Annotate the distance (Euclidean distance in this case)
distance = np.sqrt((point2[0] - point1[0])**2 + (point2[1] - point1[1])**2)
plt.annotate(f"Distance = {distance:.2f}",
             xy=((point1[0] + point2[0]) / 2, (point1[1] + point2[1]) / 2),  # Midpoint of line
             xytext=(10, 10),  # Offset the text slightly
             textcoords='offset points',
             fontsize=10, color='green',
             arrowprops=dict(facecolor='black', arrowstyle='->'))

# Add labels and title
plt.title('Plot with Annotated Distance')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```
